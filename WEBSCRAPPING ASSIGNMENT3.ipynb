```python
import selenium
from selenium import webdriver 
import pandas as pd 
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings("ignore")
import time 
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.in/")
```

```python
user_input = input("Enter what to be searched")
search = driver.find_element(By.XPATH,'//div[@class="nav-search-field "]/input')
search.send_keys(user_input)
search.send_keys(Keys.ENTER)
```

```python
from selenium.webdriver.support import expected_conditions as EC

title = []
start = 0
end = 3

for page in range(start, end):
  
    a_elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "a-size-mini.a-spacing-none.a-color-base.s-line-clamp-4"))
    )

 
    for element in a_elements:
        title.append(element.text)

    
    next_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//span[@class="s-pagination-strip"]/a[3]'))
    )
    
    
    next_button.click()

    
print(len(title))
```

```python
price = []
start = 0
end = 3
for page in range(start, end):
    a = driver.find_elements(By.CLASS_NAME,'a-price')
    for i in a:
        price.append(i.text)
    next =  driver.find_elements(By.XPATH,'//span[@class="s-pagination-strip"]/a[3]')
    time.sleep(3)
    

price= price[:104]        
print(len(price))
```

```python
expected_delivery = []
start = 0
end = 3
for page in range(start, end):
    a = driver.find_elements(By.XPATH,'//div[@class="a-row s-align-children-center"]/span/span[2]')
    for i in a:
        expected_delivery.append(i.text)
    next =  driver.find_element(By.XPATH,'//span[@class="s-pagination-strip"]/a[3]')
    next.click()
    time.sleep(3)
    

expected_delivery = expected_delivery[:104]    
print(len(expected_delivery))
```

```python
rating = []
start = 0
end = 3
for page in range(start, end):
    a = driver.find_elements(By.XPATH, '//i[contains(@class, "a-icon-star-small")]')
    for i in a:
        b = i.get_attribute('textContent')
        rating.append(b)
    next =  driver.find_element(By.XPATH,'//span[@class="s-pagination-strip"]/a[3]')
    next.click()
    time.sleep(3)
rating = rating[:104] 
print(len(rating))
```

```python
import pandas as pd 
df = pd.DataFrame({"Title":title, "Rate":price, "Rating":rating, "Expected delivery":expected_delivery})
df
```

```python
import selenium
from selenium import webdriver 
import pandas as pd 
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings("ignore")
import time 
```

```python
driver = webdriver.Chrome()
driver.get('https://youtu.be/h7eYt4ZmkR8?si=iqXySIxHE6KtEMfW')
```

```python
comments = []
a = driver.find_elements(By.XPATH,'//ytd-expander[@class="style-scope ytd-comment-renderer"]/div')
for i in a:
    comments.append(i.text)
    
len(comments)
```

```python
likes = []
b = driver.find_elements(By.ID, 'vote-count-middle')
for i in b:
    likes.append(i.text)


len(likes)
```

```python
time=[]
c = driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]/a')
for i in c:
    time.append(i.text)


len(time)
```

```python
import pandas as pd 
df = pd.DataFrame({"Comments":comments, "Likes":likes, "Time posted":time})
pd.set_option('display.max_colwidth',None)
df
```
