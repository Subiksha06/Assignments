```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

driver = webdriver.Chrome()  

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

driver.get(url)

table = driver.find_element(By.CSS_SELECTOR, 'table.wikitable')

data = []

for row in table.find_elements(By.TAG_NAME, 'tr')[1:]:
    cells = row.find_elements(By.TAG_NAME, 'td')


    if len(cells) >= 5:
       
        rank = cells[0].text.strip()
        name = cells[1].text.strip()
        artist = cells[2].text.strip()
        upload_date = cells[4].text.strip()
        view_count = cells[3].text.strip().replace(',', '')  

       
        data.append([rank, name, artist, upload_date, view_count])


df = pd.DataFrame(data, columns=["Rank", "Name", "Artist", "Upload Date", "Views"])


print(df.to_string(index=False, justify='center'))


driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
import pandas as pd

driver = webdriver.Chrome()

url = "https://www.billboard.com/"
driver.get(url) 

charts = driver.find_element(By.LINK_TEXT, "CHARTS")
charts.click() 

time.sleep(10)

top100 = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/main/div[2]/div[1]/div[1]/div/div/div[1]/div/div[2]/span/a'))
)

driver.execute_script("arguments[0].scrollIntoView();", top100)

time.sleep(5)

top100.click() 

song_name=[]
a=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li/h3')
for i in a:
    song_name.append(i.text) 
    
print(len(song_name))

artist=[]
b=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[1]/span')
for i in b:
    artist.append(i.text) 
    
print(len(artist))

lastweekrank = []
c = driver.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p u-hidden@tablet"]/ul/li[3]/span')
last_week_rank_values = [element.text.strip() for element in c]


                       
peakpos = []
d=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p u-hidden@tablet"]/ul/li[4]/span')
for i in d:
    peakpos.append(i.text) 

                       
weeksonchart=[]
e=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p u-hidden@tablet"]/ul/li[5]/span')
for i in e:
    weeksonchart.append(i.text) 

df = pd.DataFrame({"Song name":song_name, "Author name":artist, "last week rank":last_week_rank_values,"Peak positon":peakpos, "Weeks on chart":weeksonchart})
df                      
            
                       

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time 
import pandas as pd

driver = webdriver.Chrome()

url = "https://www.imdb.com/list/ls095964455/"
driver.get(url) 

name = []
a=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/a')
for i in a:
    name.append(i.text)
    
year=[]
b=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/span[2]')
for i in b:
    year.append(i.text)

genre=[]
c=driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]/span[5]')
for i in c:
    genre.append(i.text)  
    
runtime=[]
d=driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]/span[3]')
for i in d:
    runtime.append(i.text)  

rating=[]
e=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]/span[2]')
for i in e:
    rating.append(i.text)  
    
voting=[]
f=driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"][3]/span[2]')
for i in f:
    voting.append(i.text)  
    
df = pd.DataFrame({"Movie name":name, "Year":year, "Genre":genre,"RunTime":runtime, "Rating":rating, "Votes":voting})
df
    

```
